# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
import re
import json

# 🛠️ TOOL 1: Calculator
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        # Clean the expression for safety, allowing only numbers, operators, and spaces
        cleaned = re.sub(r'[^0-9\+\-\*\/\(\)\.\s]', '', expression)
        if not cleaned.strip():
            return "Error in calculation"
        return str(eval(cleaned))
    except Exception:
        return "Error in calculation"

# 🛠️ TOOL 2: Keyword Extractor
def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = []
        for w in words:
            clean_word = w.strip(".,!?;:()[]\"'").lower()
            if len(clean_word) > 4 and clean_word not in keywords:
                keywords.append(clean_word)
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION
def agent(query: str):
    query_lower = query.lower()

    # 1. Math queries -> Calculator Tool
    if "calculate" in query_lower:
        # Extract the expression part after "calculate"
        match = re.search(r'calculate\s+(.*)', query, re.IGNORECASE)
        expression = match.group(1) if match else query
        result = calculator(expression)
        if result == "Error in calculation":
            return {
                "type": "error",
                "result": result
            }
        return {
            "type": "calculation",
            "result": result
        }

    # 2. Keyword extraction -> Keyword Tool
    elif "keywords" in query_lower:
        # Extract text after "keywords" / "keywords from"
        match = re.search(r'(?:extract\s+)?keywords\s+(?:from\s+)?(.*)', query, re.IGNORECASE)
        text_to_process = match.group(1) if match else query
        result = extract_keywords(text_to_process)
        return {
            "type": "keywords",
            "result": result
        }

    # 3. General queries -> Direct response
    else:
        return {
            "type": "general",
            "result": f"I received your query: '{query}'. As this is a general query, I am responding to you directly."
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", json.dumps(agent(q), indent=2))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode
while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", json.dumps(agent(user_input), indent=2))
    print("-" * 50)
````
